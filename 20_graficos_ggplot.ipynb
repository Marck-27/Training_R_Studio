# Gráficos con ggplot

## Histogramas

```r
# install.packages('ggplot2')
# install.packages('ggplot2movies') # <-- paquete que contiene dataset de peliculas
library(ggplot2)
library(ggplot2movies)
```

```r
peliculas <- movies
class(peliculas)
```

```r
str(peliculas)
```

```r
head(peliculas)
```

```r
head(peliculas[c('title','year','rating')])
```

```r
data_to_plot <- ggplot(peliculas, aes(x =rating))
#data_to_plot
```

```r
plt_hist <- data_to_plot + geom_histogram()
plt_hist
#print(plt_hist)
```

```r
plt_hist <- data_to_plot + geom_histogram(binwidth = 0.1)
plt_hist
```

```r
plt_hist <- data_to_plot + geom_histogram(binwidth = 0.5, color = "green")
plt_hist
```

```r
plt_hist <- data_to_plot + geom_histogram(binwidth = 0.2, color = "black", fill = "green")
plt_hist
```

```r
plt_hist <- data_to_plot + geom_histogram(binwidth = 0.5, color = "black", fill = "green", alpha = 0.2)
plt_hist
```

```r
plt_hist <- data_to_plot + geom_histogram(binwidth = 0.7, color = "black", fill = "red", alpha = 0.2) + xlab("Puntuación") + ylab("Frecuencia") + ggtitle("HISTOGRAMA")
plt_hist
```

## Scatter plots

```r
coches <- mtcars
class(coches)
```

```r
head(coches)
```

```r
str(coches)
```

```r
data_to_plot <- ggplot(coches, aes(x=disp , y=mpg))
#data_to_plot
```

```r
plt_scatt <- data_to_plot + geom_point()
plt_scatt
```

```r
plt_scatt <- data_to_plot + geom_point(size = 8, color = "red", alpha = 0.4)
plt_scatt
```

```r
plt_scatt <- data_to_plot + geom_point(size = 8, color = "#2775fc", alpha = 0.4)
plt_scatt
```

```r
plt_scatt <- data_to_plot + geom_point(aes(size =wt), color = "#089562", alpha = 0.8)
plt_scatt
```

```r
options(repr.plot.width=10, repr.plot.height=5) #<-- Modifica las dimensiones de los gráficos futuros
plt_scatt <- data_to_plot + geom_point(aes(color =hp), size = 10, alpha = 0.8)
plt_scatt
```

```r
plt_scatt <- data_to_plot + geom_point(aes(color =hp), size = 10, alpha = 0.5) + scale_color_gradient(low='blue', high='red')
plt_scatt
```

```r
plt_scatt <- data_to_plot + geom_point(aes(color =hp), size = 10, alpha = 0.5) + scale_color_gradient(low='blue', high='red') + coord_cartesian(xlim = c(200,300), ylim = c(10,25))
plt_scatt
```
